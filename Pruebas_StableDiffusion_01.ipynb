```python
# Cargar las librerías
import stablediffusion as sd
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Generador de imágenes con stable diffusion
# Crea un pipeline de SD.
generador = sd.GeneradorImagenes()
```

```python
archivo = 'datasets/recipes_traducido.csv'
df = pd.read_csv(archivo, encoding='ISO-8859-1')
df.head(5)
```

```python
index = np.random.randint(len(df))
prompt = '' + str(df['nombre_del_platillo'][index])  
prompt = prompt[0:512:1]
img = generador.GenerarImagen(prompt, negative_prompt='', guidance_scale=7, num_inference_steps=25)
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title(str(df['nombre_del_platillo'][index]))
plt.show()
```
